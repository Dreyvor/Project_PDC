```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
def phi(t):
    beta = 1/2.
    T = 1

    first_coeff = (4 * beta)/(np.pi * np.sqrt(T))
    a = np.cos((1 + beta) * np.pi * t / T)
    b = ((1 - beta)*np.pi) / (4 * beta) * np.sinc((1 - beta) * t / T)
    c = 1 - np.square(4 * beta * t / T)
    
    res = first_coeff * (a+b)/c
    
    return res
```

```python
t = np.linspace(start=-10, stop=10, num=1000)

plt.figure(figsize=(16,9))
plt.plot(phi(t))
plt.grid()
plt.show()
```
